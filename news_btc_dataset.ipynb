```python
# Import Libraries
import pandas as pd
import os
```

```python
# Taking dataset from local file
news_btc = r"F:\HSE\Consulting Project\btc_analysis\news_currencies_source_joinedResult.csv"
```

```python
# Load the dataset via Pandas
df = pd.read_csv(news_btc)
df
```

```python
# Filter for BTC news only
df_btc = df[df['currencies'].astype(str).str.contains("BTC", na=False)]
```

```python
# Extract only 'BTC' and rename column to 'currency'
df_btc['currency'] = df_btc['currencies'].apply(
    lambda x: 'BTC' if isinstance(x, str) and 'BTC' in x else None
)

# Drop original 'currencies' column
df_btc = df_btc.drop(columns=['currencies'])
```

```python
#  Convert 'newsDatetime' to datetime format and extract only the date
df_btc['newsDatetime'] = pd.to_datetime(df_btc['newsDatetime'], errors='coerce')
df_btc['date'] = df_btc['newsDatetime'].dt.date
```

```python
# Rename columns
df_btc = df_btc.rename(columns={
    'title': 'news_title'
})
```

```python
# Convert specific columns to integers (fill missing with 0)
numeric_cols = ['negative', 'positive', 'important', 'liked', 'disliked', 
                'lol', 'toxic', 'saved', 'comments']
df_btc[numeric_cols] = df_btc[numeric_cols].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
```

```python
# Select relevant columns only
news_btc_dataset = df_btc[['date', 'news_title'] + numeric_cols + ['currency']]
```
