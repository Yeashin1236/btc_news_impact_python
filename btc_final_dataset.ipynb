```python
# Import Libraries
import pandas as pd
import re
```

```python
# Load dataset
file_path = r'F:\HSE\Consulting Project\btc_analysis\merged_btc.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Define a function to clean and filter titles
def filter_bitcoin_titles(title):
    if pd.isnull(title):
        return None
    # Remove all symbols (keep only letters, numbers, and spaces)
    clean_title = re.sub(r'[^A-Za-z0-9\s]', '', str(title))
    # Filter for 'Bitcoin' or 'BTC' (case-insensitive)
    if re.search(r'\bBitcoin\b|\bBTC\b', clean_title, re.IGNORECASE):
        return clean_title
    return None
```

```python
# Apply the function to the news_title column
df['filtered_title'] = df['news_title'].apply(filter_bitcoin_titles)
```

```python
# Drop rows with None in filtered_title
df_filtered = df.dropna(subset=['filtered_title'])
```

```python
btc_final_dataset = df_filtered
```

```python
# Drop original 'filtered_title' column
btc_final_dataset = btc_final_dataset.drop(columns=['filtered_title'])
```
